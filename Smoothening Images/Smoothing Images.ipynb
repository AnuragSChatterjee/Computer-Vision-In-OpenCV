```python
#Doing a 2D Convolution ( Image Filtering ) using OpenCV. This library provides a function cv.filter2D() to convolve a kernel with an image
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread(r"C:\Users\Anura\Desktop\anurag photo.jpg")

kernel = np.ones((5,5),np.float32)/25
dst = cv.filter2D(img,-1,kernel)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Doing an Image Blurring (Image Smoothing) using the functions cv.blur() or cv.boxFilter() in OpenCV.
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread(r"C:\Users\Anura\Desktop\anurag photo.jpg")


blur = cv.blur(img,(5,5)) #Doing a normal blurring of the image

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blur = cv.GaussianBlur(img,(5,5),0) #Doing a Gaussian blurring of the image

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Gaussian Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blur = median = cv.medianBlur(img,5) #Doing a Median blurring of the image

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Median Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blur = cv.bilateralFilter(img,9,75,75) #Doing a Bilateral Filtering of the image

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Bilateral Filtered')
plt.xticks([]), plt.yticks([])
plt.show()
```
